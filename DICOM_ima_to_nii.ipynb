# Convert every file into .nii format

```
import os
import dicom2nifti
import dicom2nifti.settings as settings


settings.disable_validate_orthogonal()
settings.enable_resampling()
settings.set_resample_spline_interpolation_order(1)
settings.set_resample_padding(-1000)

patient_list = 'IGT_Glioma'

for patient in os.listdir('/content/'+patient_list):
  print(patient)
  for img_modality in os.listdir('/content/'+patient_list+'/'+patient):    
    dicom2nifti.convert_directory('/content/'+patient_list+'/'+patient+'/'+img_modality+'/DICOM', '/content/'+patient_list+'/'+patient+'/')         
```
